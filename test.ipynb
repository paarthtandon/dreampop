```python
def get_track_ids(playlist):
    ids = []
    for o in range(0, 3001, 100):
        ids += [item['track']['id'] for item in sp.playlist_tracks(playlist, limit=100, offset=o)['items'] if item['track'] is not None]
    return ids

ids = []
for i, result in enumerate(results):
    print("Playlist", i)
    playlist_id = result['id']
    for id in get_track_ids(playlist_id):
        ids.append(id)
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

id_chunks = list(chunks(ids, 100))
```

```python
data = []
for c in id_chunks:
    data += sp.audio_features(tracks=c)
len(data)
```
